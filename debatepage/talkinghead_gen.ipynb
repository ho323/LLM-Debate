### HTML에 삽입하기 위해 인코딩 형식 변경

```python
import os, subprocess, shlex

input_dir = "/content/facevideo"   # 원본 영상 폴더 경로
output_dir = os.path.join(input_dir, "converted")           # 변환본 저장 폴더
os.makedirs(output_dir, exist_ok=True)

for fname in os.listdir(input_dir):
    if fname.lower().endswith(".mp4"):
        in_path = os.path.join(input_dir, fname)
        out_path = os.path.join(output_dir, os.path.splitext(fname)[0] + "_h264.mp4")

        cmd = f'ffmpeg -y -i {shlex.quote(in_path)} -c:v libx264 -pix_fmt yuv420p -c:a aac -b:a 192k {shlex.quote(out_path)}'
        print("변환 중:", fname)
        subprocess.run(cmd, shell=True)

print("모든 mp4 변환 완료! →", output_dir)
```
